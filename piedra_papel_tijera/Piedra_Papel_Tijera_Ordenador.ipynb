```python
# ========================================
# JUEGO DE PIEDRA, PAPEL O TIJERA
# ========================================

from getpass import getpass
import random

def obtener_eleccion_jugador():
    """Obtiene y valida la elección del jugador"""
    print("⚠️ Tu elección NO se verá en pantalla (es secreta)")
    eleccion = getpass("Elige (piedra/papel/tijera): ").lower()
    
    while eleccion not in ["piedra", "papel", "tijera"]:
        print("❌ Opción inválida. Debes elegir: piedra, papel o tijera")
        eleccion = getpass("Elige (piedra/papel/tijera): ").lower()
    
    return eleccion

def jugar_piedra_papel_tijera():
    """Función principal del juego"""
    
    print("=" * 50)
    print("🎮 BIENVENIDO AL JUEGO DE PIEDRA, PAPEL O TIJERA 🎮")
    print("=" * 50)
    
    # Pedimos el nombre del jugador
    jugador = input("\n👤 Tu nombre: ")
    print(f"\n¡Bienvenido/a {jugador}!")
    print("🤖 Jugarás contra la COMPUTADORA")
    print("🎯 ¡El primero en llegar a 5 puntos GANA!")
    
    # Variables para contar puntos
    puntos_jugador = 0
    puntos_computadora = 0
    
    # Bucle principal - continúa hasta que alguien llegue a 5 puntos
    while puntos_jugador < 5 and puntos_computadora < 5:
        
        print("\n" + "=" * 50)
        print("NUEVA RONDA")
        print("=" * 50)
        
        # El jugador elige
        print(f"\n{jugador}, es tu turno:")
        eleccion_jugador = obtener_eleccion_jugador()
        print(f"✅ {jugador} ya eligió su jugada")
        
        # La computadora elige aleatoriamente
        opciones = ["piedra", "papel", "tijera"]
        eleccion_computadora = random.choice(opciones)
        print("✅ La computadora ya eligió su jugada")
        
        # Mostramos las elecciones
        print("\n" + "🔓 REVELANDO ELECCIONES..." + "\n")
        print(f"🎮 {jugador} eligió: {eleccion_jugador.upper()}")
        print(f"🤖 Computadora eligió: {eleccion_computadora.upper()}")
        
        # Determinamos el ganador de la ronda
        if eleccion_jugador == eleccion_computadora:
            print("\n🤝 ¡EMPATE!")
        
        elif (eleccion_jugador == "piedra" and eleccion_computadora == "tijera") or \
             (eleccion_jugador == "papel" and eleccion_computadora == "piedra") or \
             (eleccion_jugador == "tijera" and eleccion_computadora == "papel"):
            print(f"\n🎉 ¡{jugador} GANA esta ronda!")
            puntos_jugador += 1
            
            if eleccion_jugador == "piedra":
                print("💎 La piedra rompe las tijeras")
            elif eleccion_jugador == "papel":
                print("📄 El papel envuelve la piedra")
            else:
                print("✂️ Las tijeras cortan el papel")
        
        else:
            print("\n🤖 ¡La COMPUTADORA GANA esta ronda!")
            puntos_computadora += 1
            
            if eleccion_computadora == "piedra":
                print("💎 La piedra rompe las tijeras")
            elif eleccion_computadora == "papel":
                print("📄 El papel envuelve la piedra")
            else:
                print("✂️ Las tijeras cortan el papel")
        
        # Mostramos el marcador
        print("\n" + "-" * 50)
        print("📊 MARCADOR:")
        print(f"  {jugador}: {puntos_jugador} punto(s)")
        print(f"  Computadora: {puntos_computadora} punto(s)")
        print(f"  🎯 Objetivo: 5 puntos")
        print("-" * 50)
        
        # Verificamos si alguien llegó a 5 puntos
        if puntos_jugador >= 5 or puntos_computadora >= 5:
            break
        
        # Preguntamos si quiere continuar
        while True:
            continuar = input("\n¿Quieres jugar otra ronda? (si/no): ").lower()
            if continuar in ["si", "no"]:
                break
            print("❌ Por favor responde 'si' o 'no'")
        
        if continuar == "no":
            print("\n👋 ¡Gracias por jugar! Saliendo del juego...")
            return
    
    # Mostramos resultados finales
    print("\n" + "=" * 50)
    print("🏁 ¡JUEGO TERMINADO!")
    print("=" * 50)
    
    print("\n📊 RESULTADO FINAL:")
    print(f"  {jugador}: {puntos_jugador} punto(s)")
    print(f"  Computadora: {puntos_computadora} punto(s)")
    
    if puntos_jugador >= 5:
        print(f"\n🏆 ¡{jugador} ES EL CAMPEÓN! 🏆")
    else:
        print("\n🤖 ¡La COMPUTADORA ES LA CAMPEONA! 🤖")
    
    print("\n¡Gracias por jugar! 👋")


# Ejecutamos el juego
jugar_piedra_papel_tijera()
```
