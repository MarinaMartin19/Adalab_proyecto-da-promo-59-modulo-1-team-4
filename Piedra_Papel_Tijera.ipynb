```python
# ========================================
# JUEGO DE PIEDRA, PAPEL O TIJERA
# Versión con elecciones SECRETAS (getpass)
# ========================================

# Importamos getpass para ocultar lo que escriben
from getpass import getpass

print("=" * 50)
print("🎮 BIENVENIDO AL JUEGO DE PIEDRA, PAPEL O TIJERA 🎮")
print("=" * 50)

# PASO 1: Pedimos los nombres de los jugadores
jugador1 = input("\n👤 Nombre del Jugador 1: ")
jugador2 = input("👤 Nombre del Jugador 2: ")

# Saludamos a los jugadores
print(f"\n¡Bienvenida/o {jugador1}!")
print(f"¡Bienvenida/o {jugador2}!")

# PASO 2: Creamos variables para contar los puntos
puntos_jugador1 = 0  # Empieza con 0 puntos
puntos_jugador2 = 0  # Empieza con 0 puntos
empates = 0          # Contador de empates

# PASO 3: Preguntamos si quieren jugar
jugar = "si"

# PASO 4: Bucle principal del juego (se repite mientras quieran jugar)
while jugar == "si":
    
    print("\n" + "=" * 50)
    print("NUEVA RONDA")
    print("=" * 50)
    
    # PASO 5: Jugador 1 elige (SIN QUE SE VEA)
    print(f"\n{jugador1}, es tu turno:")
    print("⚠️ Tu elección NO se verá en pantalla (es secreta)")
    eleccion1 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Validamos que la elección sea correcta
    while eleccion1 not in ["piedra", "papel", "tijera"]:
        print("❌ Opción inválida. Debes elegir: piedra, papel o tijera")
        eleccion1 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Confirmamos que eligió (sin mostrar qué eligió)
    print(f"✅ {jugador1} ya eligió su jugada")
    
    # PASO 6: Jugador 2 elige (SIN QUE SE VEA)
    print(f"\n{jugador2}, es tu turno:")
    print("⚠️ Tu elección NO se verá en pantalla (es secreta)")
    eleccion2 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Validamos que la elección sea correcta
    while eleccion2 not in ["piedra", "papel", "tijera"]:
        print("❌ Opción inválida. Debes elegir: piedra, papel o tijera")
        eleccion2 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Confirmamos que eligió (sin mostrar qué eligió)
    print(f"✅ {jugador2} ya eligió su jugada")
    
    # PASO 7: Ahora SÍ mostramos las elecciones (después de que ambos eligieron)
    print("\n" + "🔓 REVELANDO ELECCIONES..." + "\n")
    print(f"🎮 {jugador1} eligió: {eleccion1.upper()}")
    print(f"🎮 {jugador2} eligió: {eleccion2.upper()}")
    
    # PASO 8: Determinamos el ganador
    
    # CASO 1: Si eligieron lo mismo = EMPATE
    if eleccion1 == eleccion2:
        print("\n🤝 ¡EMPATE!")
        empates = empates + 1  # Sumamos 1 al contador de empates
    
    # CASO 2: Gana el Jugador 1
    elif (eleccion1 == "piedra" and eleccion2 == "tijera") or \
         (eleccion1 == "papel" and eleccion2 == "piedra") or \
         (eleccion1 == "tijera" and eleccion2 == "papel"):
        print(f"\n🎉 ¡{jugador1} GANA esta ronda!")
        puntos_jugador1 = puntos_jugador1 + 1  # Sumamos 1 punto
        
        # Explicamos por qué ganó
        if eleccion1 == "piedra":
            print("💎 La piedra rompe las tijeras")
        elif eleccion1 == "papel":
            print("📄 El papel envuelve la piedra")
        else:
            print("✂️ Las tijeras cortan el papel")
    
    # CASO 3: Gana el Jugador 2
    else:
        print(f"\n🎉 ¡{jugador2} GANA esta ronda!")
        puntos_jugador2 = puntos_jugador2 + 1  # Sumamos 1 punto
        
        # Explicamos por qué ganó
        if eleccion2 == "piedra":
            print("💎 La piedra rompe las tijeras")
        elif eleccion2 == "papel":
            print("📄 El papel envuelve la piedra")
        else:
            print("✂️ Las tijeras cortan el papel")
    
    # PASO 9: Mostramos el marcador actual
    print("\n" + "-" * 50)
    print("📊 MARCADOR:")
    print(f"  {jugador1}: {puntos_jugador1} punto(s)")
    print(f"  {jugador2}: {puntos_jugador2} punto(s)")
    print(f"  Empates: {empates}")
    print("-" * 50)
    
    # PASO 10: Preguntamos si quieren seguir jugando
    jugar = input("\n¿Quieren jugar otra ronda? (si/no): ").lower()

# PASO 11: El juego terminó - Mostramos resultados finales
print("\n" + "=" * 50)
print("🏁 ¡JUEGO TERMINADO!")
print("=" * 50)

print("\n📊 RESULTADO FINAL:")
print(f"  {jugador1}: {puntos_jugador1} punto(s)")
print(f"  {jugador2}: {puntos_jugador2} punto(s)")
print(f"  Empates: {empates}")

# PASO 12: Determinamos el campeón general
if puntos_jugador1 > puntos_jugador2:
    print(f"\n🏆 ¡{jugador1} ES EL CAMPEÓN GENERAL! 🏆")
elif puntos_jugador2 > puntos_jugador1:
    print(f"\n🏆 ¡{jugador2} ES EL CAMPEÓN GENERAL! 🏆")
else:
    print("\n🤝 ¡EMPATE GENERAL! Ambos son grandes jugadores")

print("\n¡Gracias por jugar! 👋")
```
