```python
# ========================================
# JUEGO DE PIEDRA, PAPEL O TIJERA
# VersiÃ³n con elecciones SECRETAS (getpass)
# ========================================

# Importamos getpass para ocultar lo que escriben
from getpass import getpass

print("=" * 50)
print("ğŸ® BIENVENIDO AL JUEGO DE PIEDRA, PAPEL O TIJERA ğŸ®")
print("=" * 50)

# PASO 1: Pedimos los nombres de los jugadores
jugador1 = input("\nğŸ‘¤ Nombre del Jugador 1: ")
jugador2 = input("ğŸ‘¤ Nombre del Jugador 2: ")

# Saludamos a los jugadores
print(f"\nÂ¡Bienvenida/o {jugador1}!")
print(f"Â¡Bienvenida/o {jugador2}!")

# PASO 2: Creamos variables para contar los puntos
puntos_jugador1 = 0  # Empieza con 0 puntos
puntos_jugador2 = 0  # Empieza con 0 puntos
empates = 0          # Contador de empates

# PASO 3: Preguntamos si quieren jugar
jugar = "si"

# PASO 4: Bucle principal del juego (se repite mientras quieran jugar)
while jugar == "si":
    
    print("\n" + "=" * 50)
    print("NUEVA RONDA")
    print("=" * 50)
    
    # PASO 5: Jugador 1 elige (SIN QUE SE VEA)
    print(f"\n{jugador1}, es tu turno:")
    print("âš ï¸ Tu elecciÃ³n NO se verÃ¡ en pantalla (es secreta)")
    eleccion1 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Validamos que la elecciÃ³n sea correcta
    while eleccion1 not in ["piedra", "papel", "tijera"]:
        print("âŒ OpciÃ³n invÃ¡lida. Debes elegir: piedra, papel o tijera")
        eleccion1 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Confirmamos que eligiÃ³ (sin mostrar quÃ© eligiÃ³)
    print(f"âœ… {jugador1} ya eligiÃ³ su jugada")
    
    # PASO 6: Jugador 2 elige (SIN QUE SE VEA)
    print(f"\n{jugador2}, es tu turno:")
    print("âš ï¸ Tu elecciÃ³n NO se verÃ¡ en pantalla (es secreta)")
    eleccion2 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Validamos que la elecciÃ³n sea correcta
    while eleccion2 not in ["piedra", "papel", "tijera"]:
        print("âŒ OpciÃ³n invÃ¡lida. Debes elegir: piedra, papel o tijera")
        eleccion2 = getpass("Elige (piedra/papel/tijera): ").lower()
    
    # Confirmamos que eligiÃ³ (sin mostrar quÃ© eligiÃ³)
    print(f"âœ… {jugador2} ya eligiÃ³ su jugada")
    
    # PASO 7: Ahora SÃ mostramos las elecciones (despuÃ©s de que ambos eligieron)
    print("\n" + "ğŸ”“ REVELANDO ELECCIONES..." + "\n")
    print(f"ğŸ® {jugador1} eligiÃ³: {eleccion1.upper()}")
    print(f"ğŸ® {jugador2} eligiÃ³: {eleccion2.upper()}")
    
    # PASO 8: Determinamos el ganador
    
    # CASO 1: Si eligieron lo mismo = EMPATE
    if eleccion1 == eleccion2:
        print("\nğŸ¤ Â¡EMPATE!")
        empates = empates + 1  # Sumamos 1 al contador de empates
    
    # CASO 2: Gana el Jugador 1
    elif (eleccion1 == "piedra" and eleccion2 == "tijera") or \
         (eleccion1 == "papel" and eleccion2 == "piedra") or \
         (eleccion1 == "tijera" and eleccion2 == "papel"):
        print(f"\nğŸ‰ Â¡{jugador1} GANA esta ronda!")
        puntos_jugador1 = puntos_jugador1 + 1  # Sumamos 1 punto
        
        # Explicamos por quÃ© ganÃ³
        if eleccion1 == "piedra":
            print("ğŸ’ La piedra rompe las tijeras")
        elif eleccion1 == "papel":
            print("ğŸ“„ El papel envuelve la piedra")
        else:
            print("âœ‚ï¸ Las tijeras cortan el papel")
    
    # CASO 3: Gana el Jugador 2
    else:
        print(f"\nğŸ‰ Â¡{jugador2} GANA esta ronda!")
        puntos_jugador2 = puntos_jugador2 + 1  # Sumamos 1 punto
        
        # Explicamos por quÃ© ganÃ³
        if eleccion2 == "piedra":
            print("ğŸ’ La piedra rompe las tijeras")
        elif eleccion2 == "papel":
            print("ğŸ“„ El papel envuelve la piedra")
        else:
            print("âœ‚ï¸ Las tijeras cortan el papel")
    
    # PASO 9: Mostramos el marcador actual
    print("\n" + "-" * 50)
    print("ğŸ“Š MARCADOR:")
    print(f"  {jugador1}: {puntos_jugador1} punto(s)")
    print(f"  {jugador2}: {puntos_jugador2} punto(s)")
    print(f"  Empates: {empates}")
    print("-" * 50)
    
    # PASO 10: Preguntamos si quieren seguir jugando
    jugar = input("\nÂ¿Quieren jugar otra ronda? (si/no): ").lower()

# PASO 11: El juego terminÃ³ - Mostramos resultados finales
print("\n" + "=" * 50)
print("ğŸ Â¡JUEGO TERMINADO!")
print("=" * 50)

print("\nğŸ“Š RESULTADO FINAL:")
print(f"  {jugador1}: {puntos_jugador1} punto(s)")
print(f"  {jugador2}: {puntos_jugador2} punto(s)")
print(f"  Empates: {empates}")

# PASO 12: Determinamos el campeÃ³n general
if puntos_jugador1 > puntos_jugador2:
    print(f"\nğŸ† Â¡{jugador1} ES EL CAMPEÃ“N GENERAL! ğŸ†")
elif puntos_jugador2 > puntos_jugador1:
    print(f"\nğŸ† Â¡{jugador2} ES EL CAMPEÃ“N GENERAL! ğŸ†")
else:
    print("\nğŸ¤ Â¡EMPATE GENERAL! Ambos son grandes jugadores")

print("\nÂ¡Gracias por jugar! ğŸ‘‹")
```
