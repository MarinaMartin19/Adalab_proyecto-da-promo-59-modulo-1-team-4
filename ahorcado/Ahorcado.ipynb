```python
print("==== 🎮 Ahorcado: Adivina la palabra secreta (entre dos jugadores) 🎮====")

# Persona 1 ingresa la palabra secreta

palabra_secreta = input("Persona 1, escribe una palabra secreta: ").lower()

# Para que la palabra escrita no se vea: 

print("\n" * 50)

# Vamos a crear una lista de guiones, en función del largo de la palabra_secreta

progreso = ["_"] *len(palabra_secreta)


# Definimos número de intentos 

intentos = 6

# Definimos fallos, para poder mostrar el muneco

fallos = 0


# Creamos la lista con las etapas del dibujo (desde 0 fallos hasta 6 fallos)

ahorcado = [
    # O fallos: tablero vacío 

   """
   +---+
       |
       |
       |    
     =====
    
""",

    # 1 fallo: cabeza
    """
     +---+
     😢  |
         |
         |    
       =====

""",

   # 2 fallos: cuerpo


    """
    
     +---+
     😢  |
      |  |
         |    
       =====


""",

  # 3 fallos: un brazo

    """
     +---+
     😢  |
     /|  |
         |    
       =====


""",

   

    # 4 fallos: otro brazo

    """

      +---+
     😢   |
     /|\\ |
          |
        =====
""",


    # 5 fallos: una pierna

    """

       +---+
      😢   |
      /|\\ |
      /    |
         =====
""",



    # 6 fallos: muneco completo (último intento)

    """

       +---+
       😢  |
      /|\\ |
      / \\ |
         =====
""",

]


#Lista con las letras que ya fueron probadas

Letras_nombradas = []

# Bucle principal, desarrollo del juego y importamos la función "o", para que solo se vea la última etapa del ahorcado

while intentos > 0 and "_" in progreso:

    print("\n" + ahorcado[fallos])
    
    print("Palabra: ", " ".join(progreso))

    print("Intentos restantes:", intentos)

    print("letras nombradas:", ", ".join(Letras_nombradas))


    letra = input("Adivina una letra: ").lower()

    if letra in Letras_nombradas:

        print("⚠ Esa letra ya la intentaste ⚠")
        


    else:

        Letras_nombradas.append(letra)


        if letra in palabra_secreta:

           print("✅Bien, la letra está en la palabra!✅")

        # Remplazar en la lista progreso, las posiciones correctas

           for i, l in enumerate(palabra_secreta):
                if l == letra:
                   progreso[i] = letra
    

        else:

            print("❌No está. Pierdes un intento❌")
            intentos -= 1
            fallos += 1


# Fin del juego

if "_" not in progreso:

    print(" 🎉🎉Ganaste! La palabra era:", palabra_secreta)
else:
    print(ahorcado[-1])
    print("💀Te quedaste sin intentos. La palabra era:", palabra_secreta)


print("gracias por jugar")







```
