```python
# ========================================
# JUEGO DE PIEDRA, PAPEL O TIJERA
# ========================================

from getpass import getpass
import random

def obtener_eleccion_jugador():
    """Obtiene y valida la elecciÃ³n del jugador"""
    print("âš ï¸ Tu elecciÃ³n NO se verÃ¡ en pantalla (es secreta)")
    eleccion = getpass("Elige (piedra/papel/tijera): ").lower()
    
    while eleccion not in ["piedra", "papel", "tijera"]:
        print("âŒ OpciÃ³n invÃ¡lida. Debes elegir: piedra, papel o tijera")
        eleccion = getpass("Elige (piedra/papel/tijera): ").lower()
    
    return eleccion

def jugar_piedra_papel_tijera():
    """FunciÃ³n principal del juego"""
    
    print("=" * 50)
    print("ğŸ® BIENVENIDO AL JUEGO DE PIEDRA, PAPEL O TIJERA ğŸ®")
    print("=" * 50)
    
    # Pedimos el nombre del jugador
    jugador = input("\nğŸ‘¤ Tu nombre: ")
    print(f"\nÂ¡Bienvenido/a {jugador}!")
    print("ğŸ¤– JugarÃ¡s contra la COMPUTADORA")
    print("ğŸ¯ Â¡El primero en llegar a 5 puntos GANA!")
    
    # Variables para contar puntos
    puntos_jugador = 0
    puntos_computadora = 0
    
    # Bucle principal - continÃºa hasta que alguien llegue a 5 puntos
    while puntos_jugador < 5 and puntos_computadora < 5:
        
        print("\n" + "=" * 50)
        print("NUEVA RONDA")
        print("=" * 50)
        
        # El jugador elige
        print(f"\n{jugador}, es tu turno:")
        eleccion_jugador = obtener_eleccion_jugador()
        print(f"âœ… {jugador} ya eligiÃ³ su jugada")
        
        # La computadora elige aleatoriamente
        opciones = ["piedra", "papel", "tijera"]
        eleccion_computadora = random.choice(opciones)
        print("âœ… La computadora ya eligiÃ³ su jugada")
        
        # Mostramos las elecciones
        print("\n" + "ğŸ”“ REVELANDO ELECCIONES..." + "\n")
        print(f"ğŸ® {jugador} eligiÃ³: {eleccion_jugador.upper()}")
        print(f"ğŸ¤– Computadora eligiÃ³: {eleccion_computadora.upper()}")
        
        # Determinamos el ganador de la ronda
        if eleccion_jugador == eleccion_computadora:
            print("\nğŸ¤ Â¡EMPATE!")
        
        elif (eleccion_jugador == "piedra" and eleccion_computadora == "tijera") or \
             (eleccion_jugador == "papel" and eleccion_computadora == "piedra") or \
             (eleccion_jugador == "tijera" and eleccion_computadora == "papel"):
            print(f"\nğŸ‰ Â¡{jugador} GANA esta ronda!")
            puntos_jugador += 1
            
            if eleccion_jugador == "piedra":
                print("ğŸ’ La piedra rompe las tijeras")
            elif eleccion_jugador == "papel":
                print("ğŸ“„ El papel envuelve la piedra")
            else:
                print("âœ‚ï¸ Las tijeras cortan el papel")
        
        else:
            print("\nğŸ¤– Â¡La COMPUTADORA GANA esta ronda!")
            puntos_computadora += 1
            
            if eleccion_computadora == "piedra":
                print("ğŸ’ La piedra rompe las tijeras")
            elif eleccion_computadora == "papel":
                print("ğŸ“„ El papel envuelve la piedra")
            else:
                print("âœ‚ï¸ Las tijeras cortan el papel")
        
        # Mostramos el marcador
        print("\n" + "-" * 50)
        print("ğŸ“Š MARCADOR:")
        print(f"  {jugador}: {puntos_jugador} punto(s)")
        print(f"  Computadora: {puntos_computadora} punto(s)")
        print(f"  ğŸ¯ Objetivo: 5 puntos")
        print("-" * 50)
        
        # Verificamos si alguien llegÃ³ a 5 puntos
        if puntos_jugador >= 5 or puntos_computadora >= 5:
            break
        
        # Preguntamos si quiere continuar
        while True:
            continuar = input("\nÂ¿Quieres jugar otra ronda? (si/no): ").lower()
            if continuar in ["si", "no"]:
                break
            print("âŒ Por favor responde 'si' o 'no'")
        
        if continuar == "no":
            print("\nğŸ‘‹ Â¡Gracias por jugar! Saliendo del juego...")
            return
    
    # Mostramos resultados finales
    print("\n" + "=" * 50)
    print("ğŸ Â¡JUEGO TERMINADO!")
    print("=" * 50)
    
    print("\nğŸ“Š RESULTADO FINAL:")
    print(f"  {jugador}: {puntos_jugador} punto(s)")
    print(f"  Computadora: {puntos_computadora} punto(s)")
    
    if puntos_jugador >= 5:
        print(f"\nğŸ† Â¡{jugador} ES EL CAMPEÃ“N! ğŸ†")
    else:
        print("\nğŸ¤– Â¡La COMPUTADORA ES LA CAMPEONA! ğŸ¤–")
    
    print("\nÂ¡Gracias por jugar! ğŸ‘‹")


# Ejecutamos el juego
jugar_piedra_papel_tijera()
```
